Import and parse my FourSquare Location Data (Places API) JSON file into a Pandas Data Frame
FourSquare Data - Computer Repair or Services
Manual importation - use Foursquare API from command line, create JSON file, import into labs.cognitiveclass.ai Jupyter Lab

```python
""" Import 'Computer' FourSquare Location Data into JSON file, parse location data into Pandas Data Frame / manual processing"""
# Java Script Object Notation file format
import json

with open('COMPUTER.json', 'r') as myfile:
    data = myfile.read()

obj = json.loads(data)

newobj = (json.dumps(obj, indent=4, sort_keys=True))    # convert Python Dict to String

linestring = list()
for line in newobj.splitlines():    # parse our single string into individual lines
    line2 = line.strip(" ")         # remove spaces from each string line
    line3 = line2.strip('"')        # remove the leading quotation from each line
    linestring.append(line3)        # move new line into a list item
    
""" isolate our JSON location name values """
def extractname(lines, value):

    datacleanse = list()
    for items in lines:            # strip each line of spaces
        if items.startswith(value):
            indx1 = items.index(":") + 3     # remove the leading string data
            items2 = items[indx1:] 
            indx2 = items2.index('"')         # remove the trailing string data
            items3 = items2[:indx2]
            datacleanse.append(items3)      # append each string to our list
    return datacleanse

value = "name"
names = (extractname(linestring, value))
```

```python
""" isolate our JSON location latitude values """
def extractlatitude(lines, value):

    datacleanse = list()
    for items in lines:                     # strip each line of spaces
        if items.startswith(value):
            indx1 = items.index(":") + 2    # remove the leading string data
            items2 = items[indx1:] 
            indx2 = items2.index(',')       # remove the trailing string data
            items3 = items2[:indx2]
            datacleanse.append(items3)      # append each string to our list
    return datacleanse

# extract our business competition latitude values
value = "lat"
latitude = (extractlatitude(linestring, value))
```

```python
""" isolate our computer repair business location longitude values """
def extractlongitude(lines, value):

    datacleanse = list()
    for items in lines:                      # strip each line of spaces
        if items.startswith(value):
            indx1 = items.index(":") + 2     # remove the leading string data
            items2 = items[indx1:] 
            items3 = items2[:(len(items2)-1)]   # was -1
            datacleanse.append(items3)       # append each string to our list
    return datacleanse

# extract our business competition longitude values
value = "lng"
longitude = (extractlongitude(linestring, value))
```

```python
""" consolidate our name, lat, and lng data into records, then one complete file """
tablelength = len(names)                # iterate thru the size of our parsed data
computerrepair = list()
value1 = list()                         # list [name, lat, lng]
computerplaces = list()
for items in range(0, tablelength):
    value1 = [names[items], float(latitude[items]), float(longitude[items])]
    computerplaces.append(value1)

print(computerplaces)
```

```python
""" convert our data file 'computerplaces' into a Pandas data frame """

computerplace_df = pd.DataFrame({'Computer Repair Businesses':names})    # create the base dataframe
computerplace_df['Latitude'] = latitude[0:32]                            # add latitude
computerplace_df['Longitude'] = longitude[0:32]                          # add longitude

display(computerplace_df)
```
